### Pregled predučitanih stilova

```
N = 13
cols = 2
rows = int(math.ceil(N / cols))
gs = gridspec.GridSpec(rows, cols)
fig = plt.figure(figsize=(25,20))

for style_num in range(N):
  strokes = np.load('styles/style-{}-strokes.npy'.format(style_num))
  cum = np.cumsum(strokes[:, 0:2], axis=0)
  cum = (cum - np.mean(cum, axis=0)) / np.std(cum, axis=0)
  
  ax = fig.add_subplot(gs[style_num])
  ax.scatter(cum[:, 0], cum[:, 1])
  ax.set_title('Stil {}'.format(style_num), fontsize=15)
  ax.set_xticks([])
  ax.set_yticks([])

fig.tight_layout()
```

```
import cairosvg
import os.path
from subprocess import call


lista_fileova = []
i = 0
while os.path.isfile('img/generiran_rukopis_pg{}.svg'.format(i)):
    cairosvg.svg2pdf(url='img/generiran_rukopis_pg{}.svg'.format(i), write_to='generiran_rukopis_pg{}.pdf'.format(i))
    lista_fileova.append('generiran_rukopis_pg{}.pdf'.format(i))
    call(['rm', 'img/generiran_rukopis_pg{}.svg'.format(i)])
    i += 1
```

```
from PyPDF2 import PdfFileMerger

merger = PdfFileMerger()

for pdf in lista_fileova:
    merger.append(pdf)

merger.write("generiran_rukopis.pdf")
merger.close()
```
